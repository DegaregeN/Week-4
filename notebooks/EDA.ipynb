```python
import pandas as pd

def fix_missing_value(df, columns, value):
    """
    Impute missing values in specified columns of the DataFrame with a given value.

    Parameters:
    df (DataFrame): The DataFrame in which to impute missing values.
    columns (list): List of column names to check for missing values.
    value (any): The value to replace missing entries with.

    Returns:
    DataFrame: The DataFrame with missing values imputed.
    """
    for column in columns:
        if column in df.columns:
            df[column].fillna(value, inplace=True)
    return df
```

```python
# Assuming store_df is your DataFrame
store_df = fix_missing_value(store_df, ['Promo2SinceWeek', 'Promo2SinceYear'], 0)

# Optionally, you can check the number of missing values after imputation
missing_after_imputation = store_df[['Promo2SinceWeek', 'Promo2SinceYear']].isnull().sum()
missing_after_imputation
```
