```python
import numpy as np
import pandas as pd
import sys, os
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Assuming the file is in a folder named 'data' inside 'My Drive'
file_path = '/content/drive/My Drive/data/merged_store.csv'

# Load the file into a DataFrame
df = pd.read_csv(file_path)
df
```

```python
df.describe()
```

```python
# unique value counts of store type

counts_df = df['StoreType'].value_counts()
counts_df
```

```python
# Unique value counts of 'StoreType' column
counts_df = df['StoreType'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.bar(counts_df.index, counts_df.values, color=['lightblue', 'lightgreen', 'lightcoral', 'orange'])

# Add title and labels
plt.title('Distribution of Store Types', fontsize=16)
plt.xlabel('Store Type', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Display the chart
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.show()
```

```python
# unique value counts
counts_df = df['Assortment'].value_counts()
counts_df
```

```python
# Unique value counts of 'StoreType' column
counts_df = df['Assortment'].value_counts()

# Create a pie chart
plt.figure(figsize=(6,6))  # Set the figure size
plt.pie(counts_df, labels=counts_df.index, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])

# Add a title
plt.title('Distribution of Assortment')

# Display the chart
plt.show()
```

```python
# Plot a histogram for the 'CompetitionDistance' column
plt.figure(figsize=(10,6))  # Set figure size
sns.histplot(df['CompetitionDistance'], bins=30, kde=False, color='blue')

# Add labels and title
plt.title('Distribution of Competition Distance', fontsize=16)
plt.xlabel('Competition Distance', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
# unique value counts
counts_df = df['Promo2'].value_counts()
counts_df
```

```python
# unique value counts
counts_df = df['CompetitionBeforeStoreOpened'].value_counts()
counts_df
```

```python
# Plot a barplot for 'DayOfWeek' without specifying a palette
plt.figure(figsize=(10,6))  # Set figure size
sns.countplot(x='DayOfWeek', data=df)

# Add labels and title
plt.title('Count of Sales by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```

```python
# Plot a histogram for the 'Sales' column
plt.figure(figsize=(10,6))  # Set figure size
sns.histplot(df['Sales'], bins=30, kde=False, color='blue')

# Add labels and title
plt.title('Distribution of Sales', fontsize=16)
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
# Plot a histogram for the 'Sales' column
plt.figure(figsize=(10,6))  # Set figure size
sns.histplot(df['Customers'], bins=30, kde=False, color='skyblue')

# Add labels and title
plt.title('Distribution of Customers', fontsize=16)
plt.xlabel('Customers', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
# Plot a barplot for 'Open' column
plt.figure(figsize=(8,6))  # Set figure size
sns.countplot(x='Open', data=df, color='skyblue')

# Add labels and title
plt.title('Count of Open vs Closed Stores', fontsize=16)
plt.xlabel('Open (1 = Open, 0 = Closed)', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```

```python
# Plot a barplot for 'Promo' column
plt.figure(figsize=(8, 6))  # Set figure size
sns.countplot(x='Promo', data=df, color='lightcoral')

# Add labels and title
plt.title('Count of Promo vs No Promo', fontsize=16)
plt.xlabel('Promo (1 = Active, 0 = Inactive)', fontsize=12)
plt.ylabel('Count', fontsize=12)

# Show the plot
plt.show()
```
